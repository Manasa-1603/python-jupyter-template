### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

x_line=np.linspace(0,5,50)
y_line=x_line**2

categories_bar=['X','Y','Z']
values_bar=[10,15,7]

x_scatter=[1,2,3,4,5]
y_scatter=[3,1,4,2,5]

labels_pie=['Part 1','Part 2','Part 3']
sizes_pie=[45,30,25]
colors_pie=['gold','silver','lightcoral']
explode_pie=(0.1,0,0)

fig,axes=plt.subplots(2,2,figsize=(10,8))

if not x_line.size or not y_line.size:
    print('Error (Subplot 1): X or Y line data is empty.')
else:
    axes[0,0].plot(x_line,y_line)
    axes[0,0].set_title('Line chart')
    axes[0,0].set_xlabel('X')
    axes[0,0].set_ylabel('Y')
    axes[0,0].grid(True)
if len(categories_bar) != len(values_bar):
    print("Error (Subplot 2): Inconsistent lengths for bar chart categories and values.")
else:
    axes[0,1].bar(categories_bar,values_bar)
    axes[0,1].set_title('Bar chart')
    axes[0,1].set_xlabel('X')
    axes[0,1].set_ylabel('Y')
if len(x_scatter) != len(y_scatter):
    print("Error (Subplot 3): Inconsistent lengths for scatter plot x and y data.")
else:
    axes[1,0].scatter(x_scatter,y_scatter)
    axes[1,0].set_title('Scatter plot')
    axes[1,0].set_xlabel('X')
    axes[1,0].set_ylabel('Y')
    axes[1,0].grid(True)
if len(labels_pie) != len(sizes_pie):
    print("Error (Subplot 4): Inconsistent lengths for pie chart labels and sizes.")
elif sum(sizes_pie) != 100:
    print("Warning (Subplot 4): The sum of pie chart sizes is not 100.")
else:
    axes[1,1].pie(sizes_pie,explode=explode_pie,labels=labels_pie,colors=colors_pie,
                autopct='%1.1f%%',shadow=True,startangle=140)
    axes[1,1].axis('equal')
    axes[1,1].set_title('Pie chart')

plt.tight_layout()
plt.show()
```
