### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

data=[1,2,2,3,3,3,4,4,4,4,5]
num_bins = 5
if num_bins <= 0:
    print("Error: The number of bins must be a positive integer.")
elif num_bins > len(set(data)):
    print("Warning: The desired number of bins exceeds the number of unique data points. This might lead to empty bins.")
else:
    plt.hist(data,bins=num_bins,edgecolor='black',color='skyblue')
    plt.title('Histogram of data distribution')
    plt.xlabel('Data values')
    plt.ylabel('Frequency')
    plt.show()
```
