### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]

if len(x)!=len(y):
    print('Error:The lengrh of x and y data must be the same for a scatter plot.')
else:
    plt.scatter(x,y)
    for i in range(len(x)):
        plt.annotate(f'({x[i]},{y[i]})',(x[i],y[i]),
                    textcoords='offset points',
                    xytext=(5,5),
                    ha='left')
    
    plt.title('Scatter plot with annotations')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()
```
